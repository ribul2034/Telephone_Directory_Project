```python
def add_new_contact():
    global df
#     df = df
    contact_number=input('\tEnter the new contact number: ')
    if(len(contact_number))>9 and contact_number.isnumeric()==True:
        for i in range(len(df)):
            if df['Contact_number'][i]==contact_number:
                return "\033[31m"+'\tThis number is already present in the telephone directory as ' + df['Contact_name'][i]+"\033[0m"


        contact_name = input("\tEnter the contact name: ")
        if len(contact_name)<1:
            df = df.append({'Contact_number':contact_number, 'Contact_name':" "}, ignore_index=True)
            df.to_csv('data.csv', index=False)
            return "\033[32m" + "\tContact saved successfully"+"\033[0m"
        
        for i in range(len(df)):
            if df['Contact_name'][i]==contact_name:
                return "\033[31m"+ "\tWarning! This name is already present in Telephone Directory"+ "\033[0m"

        df = df.append({'Contact_number':contact_number, 'Contact_name':contact_name}, ignore_index=True)
        df.to_csv('data.csv', index=False)
        return "\033[32m" + "\tContact saved successfully"+ "\033[0m"
    else:
        print("\033[31m"+"\tPlease provide valid contact number"+"\033[0m")
        add_new_contact()
```

```python
def search_contact():
    global df
    search = input("Enter the data for searching: ")
    print(search)
    search_df=pd.DataFrame(columns=['Contact_number', 'Contact_name'])
    for i in range(len(df)):
        if search in df['Contact_number'][i] or search in df['Contact_name'][i]:
            search_df=search_df.append({'Contact_number':df['Contact_number'][i], 'Contact_name':df['Contact_name'][i]},ignore_index=True)
            
    if len(search_df)==0:
        return "\033[31m"+"\tSearch data is not found"+"\033[0m"
    else:
        print(search_df.style.applymap(lambda x: 'color: blue'))
        return "\033[32m" +"\tData is searched successfully"+"\033[0m"
    
```

```python
def update_contact():
    global df
    search=input("\tEnter the detail for updatation: ")
    if len(df) != 0:
        for i in range(len(df)):
            if search==df['Contact_number'][i] or search==df['Contact_name'][i]:
                print("\tContact number is "+ "\033[34m" +df['Contact_number'][i]+ "\033[0m"+ " and Contact name is " + "\033[34m" +df['Contact_name'][i]+ "\033[0m")
                user_choice=input('\tIs you want to edit the above contact detail(yes/no): ')
                if user_choice=='yes':
                    user_choice1=input("""\tWhat you want to change: 
                    \t1. For name(Press 1)
                    \t2. For number(Press 2)""")
                    if user_choice1=='1':
                        contact_name2=input("\tContact name: ")
                        df['Contact_name'][i]=contact_name2
                    elif user_choice1=='2':
                        contact_number2=input("\tContact number: ")
                        if(len(contact_number))>9 and contact_number.isnumeric()==True:
                            df['Contact_number'][i]=contact_number2
                        else:
                            return "\033[31m"+"Please provide valid contact number"+"\033[0m"
                    else:
                        return "\033[34m"+"\tNo updation"+"\033[0m"
                    df.to_csv('data.csv', index=False)
                    return "\033[32m"+'\tContact detail is updated successfully' +"\033[0m"
            else:
                return "\033[34m"+"\tNo updation"+"\033[0m"
    return "\033[31m"+"\tInput detail is not found"+"\033[0m"
    
```

```python
def delete_contact():
    global df
    search = input("\tEnter the data for deletion: ")
    if len(df) != 0:
        for i in range(len(df)):
            if search==df['Contact_number'][i] or search==df['Contact_name'][i]:
                print("\tContact number is "+ "\033[34m"+ {df['Contact_number'][i]} + "\033[0m" +" and Contact name is "+"\033[34m"+ {df['Contact_name'][i]} +"\033[0m")
                user=input('\tIs you want to delete the above contact number(yes/no): ')
                if user=='yes':
                    df=df.drop(df.index[1], inplace=True)
                    df.to_csv('data.csv', index=False)
                    return "\033[32m"+"\tThe contact has been deleted successfully!!!"+"\033[0m"
                else:
                    return "\033[34m"+"\tThe contact isn't deleted!!!"+"\033[0m"
    return "\033[31m"+"\tThe contact detail not found"+"\033[0m"
        
    
```

```python
def export():
    from datetime import datetime
    s=datetime.today()
    s=s.strftime("%d_%m_%y")
    s="Telephone_Directory_"+s+".csv"
    
    df.to_csv(s)
    return "\033[32m" + "\tContacts exported successfully" + "\033[0m"
    
```

```python
def menu_function():
    
    print(Fore.BLUE +"\t Choose option from given list: " + Style.RESET_ALL)
    menu_text = """\t1. Add new contact
    \t2. Search contact
    \t3. Update the existing contact
    \t4. Delete the contact
    \t5. Export the contact
    \t6. Exit"""
    print(menu_text)
    user_choice=input("\tSelect the option from above: ")
    if user_choice=='1':
        a=add_new_contact()
        print(a)
        print()
        print("-"*100)
        menu_function()
    elif user_choice=='2':
        a=search_contact()
        print(a)
        print()
        print("-"*100)
        menu_function()
    elif user_choice=='3':
        a=update_contact()
        print(a)
        print()
        print("-"*100)
        menu_function()
    elif user_choice=='4':
        a=delete_contact()
        print(a)
        print()
        print("-"*100)
        menu_function()
    elif user_choice=='5':
        a=export()
        print(a)
        print()
        print("-"*100)
        menu_function()
    elif user_choice=='6':
        return print("\033[32m"+"\tExit from telephone directory"+"\033[0m")
    else:
        print("\033[31m" +"Please provide appropriate choice!!!!"+"\033[0m")
        menu_function()
```
